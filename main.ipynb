```python
from PIL import Image
import os

main_path = 'FamousPeopleBD'

for folder in os.listdir(main_path):
    folder_path = os.path.join(main_path, folder)
    if os.path.isdir(folder_path):
        for file in os.listdir(folder_path):
            file_path = os.path.join(folder_path, file)
            try:
                # Try opening as image
                im = Image.open(file_path)
                im = im.convert("RGB")  # ensure proper format

                # Save as PNG
                new_file = os.path.splitext(file_path)[0] + ".png"
                im.save(new_file)

                # Remove original if not already .png
                if not file.lower().endswith(".png"):
                    os.remove(file_path)
                    print("Converted:", file_path, "→", new_file)

            except Exception as e:
                print("Skipping corrupted file:", file_path, "| Error:", e)
```

```python
import tensorflow as tf

dataset = tf.keras.preprocessing.image_dataset_from_directory(
    main_path,      # main folder path
    image_size=(256, 256), # resize images
    batch_size=32,         # how many per batch
    labels="inferred",     # use folder names as labels
    shuffle=True
)
```

```python
def preprocess(image, label):
    image = tf.cast(image, tf.float32) / 255.0
    return image, label

train_ds = dataset.map(preprocess)
```

```python
import os

main_path = 'FamousPeopleBD'
class_names = sorted(os.listdir(main_path))  # get folder names as classes
num_classes = len(class_names)

print(num_classes)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models


model = models.Sequential([
    layers.InputLayer(input_shape=(256, 256, 3)),

    # Convolutional layers
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(128, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),

    # Fully connected layers
    layers.Dense(128, activation='relu'),
    layers.Dense(num_classes, activation='softmax')  # output layer
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # use sparse because labels are integers
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_ds,
    epochs=10  # you can increase later
)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import os

# --- Parameters ---
img_path = "download.jpeg"  # replace with your image path
main_path = "FamousPeopleBD"          # your dataset folder (for class names)

# --- Load class names ---
class_names = sorted([d for d in os.listdir(main_path) if os.path.isdir(os.path.join(main_path, d))])

# --- Load and preprocess image ---
img = image.load_img(img_path, target_size=(256, 256))  # resize to model input
img = img.convert("RGB")  # ensure 3 channels
img_array = image.img_to_array(img) / 255.0             # scale 0-1
img_array = np.expand_dims(img_array, axis=0)          # add batch dimension

# --- Predict ---
predictions = model.predict(img_array)                 # model prediction
predicted_index = np.argmax(predictions[0])            # index of highest probability
predicted_class = class_names[predicted_index]        # get class name
confidence = predictions[0][predicted_index]

print(f"Predicted: {predicted_class} (confidence: {confidence:.2f})")
```
